```python
import csv
import numpy as np
from clusterone import ClusteroneClient
# Load PPI network data
network = np.loadtxt('ppi_network.txt', dtype=int)
# Load known protein complex data
with open('known_complexes.txt', 'r') as f:
    reader = csv.reader(f, delimiter=' ')
    complexes = []
    for row in reader:
        # Ignore any extra fields in the row
        complex_ids = [int(i) for i in row if i]
        complexes.append(complex_ids)
# Create ClusterOneClient object with default parameters
client = ClusteroneClient()
# Find communities in the PPI network using the ClusterONE algorithm
communities = client.run(network, complexes)
# Print the identified protein complexes
for i, community in enumerate(communities):
    print(f'Community {i+1}: {community}')
```
